```python
# Create histogram for Percent Households below Poverty field 
plt.hist(df['PERCENT HOUSEHOLDS BELOW POVERTY'], bins=5, color='#33AFFF', edgecolor='black')

# Add labels and title
plt.xlabel('Percent Households Below Poverty')
plt.ylabel('Frequency')
plt.title('Frequency of Percent Households Below Poverty')

# Show the plot
plt.show()
```

```python
# Create Pairplot
sns.pairplot(df)
plt.show()
```

```python
# Create Correlation Coefficient Matrix
correlation_matrix = df[["PERCENT HOUSEHOLDS BELOW POVERTY", "PERCENT OF HOUSING CROWDED", "PERCENT AGED 16+ UNEMPLOYED", "PERCENT AGED 25+ WITHOUT HIGH SCHOOL DIPLOMA", "PERCENT AGED UNDER 18 OR OVER 64", "PER CAPITA INCOME"]].corr()
correlation_matrix
```

```python
# Create Heatmap 
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f", square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Create scatter plot for Percent Aged 15+ Without High School Diploma and Percent Households Below Povery
scatter_plot = df.plot(kind="scatter", x="PERCENT AGED 25+ WITHOUT HIGH SCHOOL DIPLOMA", y="PERCENT HOUSEHOLDS BELOW POVERTY", grid=True, figsize=(10,8), title= "High School Diploma vs Poverty Level", color='#33AFFF')

# Create x and y values 
x_values=df['PERCENT AGED 25+ WITHOUT HIGH SCHOOL DIPLOMA']
y_values=df['PERCENT HOUSEHOLDS BELOW POVERTY']

# Calculate linear regression 
(slope, intercept,  rvalue, pvalue, stderr)= st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
plt.plot(x_values,regress_values,'#3633FF')
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(f'Linear regression equation: {line_eq}')

# Show plot 
plt.show()
```
